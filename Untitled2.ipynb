```python
import os, warnings
warnings.filterwarnings("ignore")
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import SpectralClustering, KMeans, AgglomerativeClustering, DBSCAN
from sklearn.metrics import silhouette_score
```

```python
DATA_PATH = "/content/Spotify_data.xlsx"
df = pd.read_excel(DATA_PATH)
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if len(numeric_cols) == 0:
    for col in df.columns:
        try:
            df[col] = pd.to_numeric(df[col], errors="ignore")
        except Exception:
            pass
    numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
if len(numeric_cols) == 0:
    raise ValueError("No numeric columns found. Add features like danceability, energy, tempo, loudness, valence, etc.")

X = df[numeric_cols].copy()

for c in X.columns:
    if X[c].isna().any():
        X[c].fillna(X[c].median(), inplace=True)
const_cols = X.columns[X.nunique() <= 1].tolist()
if const_cols:
    X.drop(columns=const_cols, inplace=True)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca_n = min(2, max(1, X.shape[1]))
pca = PCA(n_components=pca_n, random_state=42)
X_pca = pca.fit_transform(X_scaled)

if X.shape[1] == 1:
    s_k = km_k = ag_k = 3
else:
    def best_k_search(fit_fn, k_min=2, k_max=6):
        best_k, best_score = None, -1
        for k in range(k_min, min(k_max, X_scaled.shape[0]-1)+1):
            try:
                labels = fit_fn(k)
                if len(set(labels)) <= 1:
                    continue
                score = silhouette_score(X_scaled, labels)
                if score > best_score:
                    best_score, best_k = score, k
            except Exception:
                continue
        return best_k if best_k is not None else 3

    s_k = best_k_search(lambda k: SpectralClustering(n_clusters=k, affinity='nearest_neighbors', n_neighbors=min(10, X_scaled.shape[0]-1), random_state=42, assign_labels='kmeans').fit_predict(X_scaled))
    km_k = best_k_search(lambda k: KMeans(n_clusters=k, random_state=42, n_init=10).fit_predict(X_scaled))
    ag_k = best_k_search(lambda k: AgglomerativeClustering(n_clusters=k).fit_predict(X_scaled))
```

```python
spectral_labels = SpectralClustering(n_clusters=s_k, affinity='nearest_neighbors', n_neighbors=min(10, X_scaled.shape[0]-1), random_state=42, assign_labels='kmeans').fit_predict(X_scaled)
kmeans_labels = KMeans(n_clusters=km_k, random_state=42, n_init=10).fit_predict(X_scaled)
agglom_labels = AgglomerativeClustering(n_clusters=ag_k).fit_predict(X_scaled)

best_dbscan = {"eps": None, "min_samples": None, "score": -1, "labels": None}
for eps in [0.3, 0.5, 0.7, 1.0]:
    for ms in [3,5,8]:
        labels_db = DBSCAN(eps=eps, min_samples=ms).fit_predict(X_scaled)
        ncl = len(set(labels_db)) - (1 if -1 in labels_db else 0)
        if ncl <= 1:
            continue
        try:
            score = silhouette_score(X_scaled, labels_db)
        except Exception:
            score = -1
        if score > best_dbscan["score"]:
            best_dbscan.update({"eps": eps, "min_samples": ms, "score": score, "labels": labels_db})
dbscan_labels = best_dbscan["labels"] if best_dbscan["labels"] is not None else np.array([-1]*X_scaled.shape[0])
```

```python
out_df = df.copy().reset_index(drop=True)
out_df["spectral_cluster"] = spectral_labels
out_df["kmeans_cluster"] = kmeans_labels
out_df["agglom_cluster"] = agglom_labels
out_df["dbscan_cluster"] = dbscan_labels

OUT_PATH = "/content/Spotify_clusters.csv"
out_df.to_csv(OUT_PATH, index=False)
print("Saved clustered CSV to:", OUT_PATH)

def plot_clusters(points, labels, title):
    plt.figure(figsize=(8,5))
    if points.shape[1] == 1:
        jitter = np.random.normal(scale=0.02, size=points.shape[0])
        plt.scatter(points[:,0], jitter, c=labels, s=30)
        plt.yticks([])
        plt.xlabel("PCA 1 (1D)")
    else:
        plt.scatter(points[:,0], points[:,1], c=labels, s=20)
        plt.xlabel("PCA 1"); plt.ylabel("PCA 2")
    plt.title(title); plt.grid(True); plt.tight_layout(); plt.show()

plot_clusters(X_pca, spectral_labels, f"Spectral (k={s_k})")
plot_clusters(X_pca, kmeans_labels, f"KMeans (k={km_k})")
plot_clusters(X_pca, agglom_labels, f"Agglomerative (k={ag_k})")
if best_dbscan["labels"] is not None:
    plot_clusters(X_pca, dbscan_labels, f"DBSCAN (eps={best_dbscan['eps']}, min_samples={best_dbscan['min_samples']})")

def safe_silhouette(lbls):
    try:
        if len(set(lbls)) <= 1 or (len(set(lbls))==2 and -1 in set(lbls)):
            return None
        return silhouette_score(X_scaled, lbls)
    except:
        return None

print("Silhouette scores:", {
    "spectral": safe_silhouette(spectral_labels),
    "kmeans": safe_silhouette(kmeans_labels),
    "agglomerative": safe_silhouette(agglom_labels),
    "dbscan": safe_silhouette(dbscan_labels)
})
```
